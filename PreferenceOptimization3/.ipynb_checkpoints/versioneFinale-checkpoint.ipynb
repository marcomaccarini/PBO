```python
import matplotlib.pyplot as plt
import GLISp as GL
import model.RBF_model as rbf
import numpy as np
from preference_functions import compute_preference, compute_class
```

```python
def my_fun(X, **kwarg):
    ''' f(x) = (1+(xsin(2x)cos(3x))/(1+x^2))^2+x^2/12+x/10'''
    fval = (1 + (X[:, 0] * np.sin(2 * X[:, 0]) * np.cos(3 * X[:, 0])) /
            (1 + X[:, 0] ** 2)) ** 2 + X[:,0] ** 2 / 12 + X[:,0] / 10
    return fval

def my_fun_easy(X, **kwarg):
    ''' f(x) = (1+(xsin(2x)cos(3x))/(1+x^2))^2+x^2/12+x/10'''
    fval = (1 + (X * np.sin(2 * X) * np.cos(3 *X )) /
            (1 + X ** 2)) ** 2 + X ** 2 / 12 + X/ 10
    return fval
```

```python
def f_class_1(X):
    Y_class = (X[:, 0] <= 10) & (X[:, 0] >= -10)
    return Y_class


def f_class_2(X):
    Y_class = (X[:, 0] <= 10) & (X[:, 0] >= -10)
    return Y_class
```

```python
def my_pref(x1, x2, f):
    print(x1)
    Y_pref = compute_preference(f, x1, x2)
    Y_pref.astype(int)
    print("function: ", f)
    print("x1: ", x1,"x2: ", x2, "f(x1): ", f(x1) ,"f(x2): ", f(x2) , "result: ", Y_pref )
    return Y_pref
```

```python
bb = 3

fvarsX = [{'name': 'x1', 'type': 'continuous', 'domain': (-bb, bb)}]

n = len(fvarsX)

```

```python
X_class = np.random.rand(30, n) * (2 * bb) - bb
y_class = f_class_1(X_class)

X_class2 = np.random.rand(30, n) * (2 * bb) - bb
y_class2 = f_class_2(X_class2)

```

```python
GP_class = {'X': [X_class, X_class2], 'Y': [y_class, y_class2]}

init_samples = 10  # Number of initial samples

my_model = rbf.RBF_model(fvarsX, normalize_X=True, plot=True, save= False)#, theta=1/5, lam=0)


```

```python
import utils
```

```python
my_problem = GL.GLISp(my_fun, my_model, my_pref, acquisition_optimizer_type='lbfgsb', \
                      kfold=3, delta=0.5, objective=(10, 1), \
                      batch_size=1, my_fun=my_fun, save_experiment=False,plot = False, save = False, load_experiment_foldername = 'Experiment20210503-185211')

#se il k-fold è None, allora fa l'ottimizzazione con il convex problem senza fare la recalibrazione dei parametri



ris = my_problem.run_optimization(max_iter=10)
```
