```python
import GlispFinale as GL
import numpy as np
from preference_functions import compute_preference
# from rosFolder import talkerPreferencesMultiArray as t
from datetime import datetime
import utils
import math
```

```python
def point_eval_out1(point):
    x1 = point[0]
    x2 = point[1]
    r  = x1**2 + x2**2
    # xopts = [[0,0]]  # unconstrained optimizers, one per column
    # fopts = 0  # unconstrained optimum
    return r
```

```python
def out1(X):
    array = []
    for x in X:
        array.append(point_eval_out1(x))
    return np.array(array)
```

```python
xopt0 = [[0,0]]  # unconstrained optimizers, one per column
Ybest = 0 # unconstrained optimum
```

```python
def my_pref(x1, x2, f):
    if f is not None:
        Y_pref = compute_preference(f, x1, x2)
        Y_pref.astype(int)
        print("best is :  %20s, next is: %20s, f(best): %4.4f, f(next): %4.4f, user's input: %d" % (
            str(x1), str(x2), f(x1), f(x2), Y_pref))
        yy.append(f(x1)[0])
        return Y_pref
    else:
        print("best is : ", x1, " , x next is : ", x2)
        Y_pref = input("1 if x_next is worse than best so far, -1 otherwise")
        return int(Y_pref)
```

```python
def g(xt, Y, Models):
    return Y[:, 0]
```

```python
bb = 10
fvarsX = [
    {'name': 'x1', 'type': 'continuous', 'domain': (-bb, bb)},
    {'name': 'x2', 'type': 'continuous', 'domain': (-bb, bb)}
]
fvarsY = [
    {'name': 'out1', 'type': 'continuous', 'domain': ()}
]
```

```python
n = len(fvarsX)

init_samples = 10  # Number of initial samples

objectives = [out1]
objectivesFunctions = [point_eval_out1]
```

```python
my_problem = GL.GLISpFinale(fvarsX, fvarsY, my_pref, objectives=objectives,
                            objectivesFunctions=objectivesFunctions,
                            acquisition_optimizer_type='lbfgsb',
                            kfold=3, delta=0.5,
                            objective=(10, 1),
                            batch_size=1,
                            save_experiment=False, plot=True,
                            g=g,
                            theta=2, sigma=1 / 5, lam=2,
                            max_bound=bb,
                            name_opt='forPaper1')
```

```python
exploration = True
x_next = my_problem.run_optimization(exploration=exploration)
iterations = 15
print("X next: ", x_next)
```

```python
xx = np.arange(0, iterations, 1)  # sono gli step
yy = []  # sono le f(best)
fyy = []  # sono le f(next)

end_explore = int(iterations * 0.1)
```

```python
for i in range(iterations):
    eval = []
    y_sum = 0
    print(i)
    for ind in range(len(my_problem.models)):
        i_best = my_problem.Y_ind_best[ind]
        eval.append(
            my_pref(
                my_problem.models[ind].X[i_best:i_best + 1, :],

                my_problem.X_next[0: 1, :],

                f=objectives[ind])
        )  # for all y there will be an evaluation
        y_sum += objectives[ind](x_next)
    fyy.append(y_sum)
    if i > iterations - end_explore:
        exploration = False
    my_problem.add_evaluations(x_next, eval)
    x_next = my_problem.run_optimization(exploration=exploration)
    # print(x_next)
```

```python
init_n_samples=1000
x_plot = np.empty((init_n_samples, 0))
for function_var in my_problem.fvars_x:
    if function_var.get_type() == 'continuous':
        x = np.random.uniform(function_var.get_domain()[0], function_var.get_domain()[1],
                              (init_n_samples, 1))
    elif function_var.get_type() == 'discrete':
        x = np.random.choice(function_var.get_domain(), (init_n_samples, 1))
    elif function_var.get_type() == 'categorical':
        x = function_var.get_one_hot(np.random.choice(function_var.get_domain(), (init_n_samples, 1)))
    else:
        raise ValueError('Variable type not supported')
    x_plot = np.column_stack((x_plot, x))
```
